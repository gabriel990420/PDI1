```
def mostrar3Grises(im1, im2, im3, title1="original", title2="Editado1", title3="Editado2"):
  plt.figure(figsize=(14,14))
  plt.subplot(131), plt.imshow(im1, cmap= 'gray'), plt.title(title1)
  plt.xticks([]), plt.yticks([])
  plt.subplot(132), plt.imshow(im2, cmap='gray'), plt.title(title2)
  plt.xticks([]), plt.yticks([])
  plt.subplot(133), plt.imshow(im3, cmap='gray'), plt.title(title3)
  plt.xticks([]), plt.yticks([])
  plt.show()
  
def mostrar2Grises(im1, im2, title1="original", title2="Editado2"):
  plt.figure(figsize=(14,14))
  plt.subplot(121), plt.imshow(im1, cmap= 'gray'), plt.title(title1)
  plt.xticks([]), plt.yticks([])
  plt.subplot(122), plt.imshow(im2, cmap='gray'), plt.title(title2)
  plt.xticks([]), plt.yticks([])
  plt.show()
  
def mostrar3Colores(im1, im2, im3, title1="original", title2="Editado1", title3="Editado2"):
  plt.figure(figsize=(14,14))
  plt.subplot(131), plt.imshow(im1), plt.title(title1)
  plt.xticks([]), plt.yticks([])
  plt.subplot(132), plt.imshow(im2), plt.title(title2)
  plt.xticks([]), plt.yticks([])
  plt.subplot(133), plt.imshow(im3), plt.title(title3)
  plt.xticks([]), plt.yticks([])
  plt.show()
  
def mostrar2Colores(im1, im2, title1="original", title2="Editado2"):
  plt.figure(figsize=(7,7))
  plt.subplot(121), plt.imshow(im1), plt.title(title1)
  plt.xticks([]), plt.yticks([])
  plt.subplot(122), plt.imshow(im2), plt.title(title2)
  plt.xticks([]), plt.yticks([])
  plt.show()
```

# **Interpolacion Vecino mas cercano**

```
file1 = path+"ico.jpg"
file2 = path+"apple20x22.jpg"
im = cv2.imread(file1)
im2 = cv2.imread(file2)
imRGB = cv2.cvtColor(im,cv2.COLOR_BGR2RGB)
imRGB2 = cv2.cvtColor(im2,cv2.COLOR_BGR2RGB)

near_imgA = cv2.resize(imRGB, None, fx = 1, fy = 1, interpolation = cv2.INTER_NEAREST)
near_imgB = cv2.resize(imRGB2, None, fx = 5, fy = 5, interpolation = cv2.INTER_NEAREST)

mostrar2Colores(near_imgA, near_imgB, "señora", "manzana")
```

```
def negative(L):
  #cargar imagen
  img = cv2.imread("ojo.jpg", cv2.IMREAD_GRAYSCALE)
  rows, columns = img.shape
  img2 = np.zeros((rows, columns), dtype= np.uint8)
  for x in range(rows):
    for y in range(columns):
      img2[x,y] = (L-1)-img[x,y]
  mostrar2Grises(img, img2, 'original', 'negativo')
  return 

negative(2**8)
```

```
img = cv2.imread('ojo.jpg')
img_log = (np.log(img+1)/(np.log(1+np.max(img))))*255
img_log = np.array(img_log, dtype = np.uint8)
mostrar2Grises(img, img_log)
```

#Binarizacion

```
def binarizacion(img, minr=10,maxr=60):
  img = cv2.imread(img,0)
  row, column = img.shape
  img1=np.zeros((row,column),dtype='uint8')
  
  # specify the min and max range
  min_range=minr
  max_range=maxr
  
  #loop over the input image and if pixel value lies in desired range 
  for i in range(row):
    for j in range(column):
      if img[i,j]>min_range and img[i,j]<max_range:
        img1[i,j]=255
      else:
        img1[i,j]=0
        
  # display the image
  
  mostrar2Grises(img,img1)
 

binarizacion('tree.png')
```

#Fraccionamiento

```
def fraccionamiento(img,minr=10,maxr=60):
  img=cv2.imread(img,cv2.IMREAD_GRAYSCALE)
  rows,columns = img.shape
  img2=np.zeros((rows,columns),dtype=np.uint8)
  
  #intervalos
  a=minr
  b=maxr
  
  for x in range(0,rows):
    for y in range(0,columns):
      r=img[x][y]
      if a <=r and r<=b:
        img2[x][y] =255
      else:
        img2[x][y]= img[x][y]
  mostrar2Grises(img,img2)
  return

#fraccionamiento('pastel.png',200,250)
binarizacion('pastel.png',10,60)
```

#Threshold

```
def threshold(img,thres):
  img = cv2.imread(img,cv2.IMREAD_GRAYSCALE)
  rows, columns = img.shape
  img2=np.zeros((rows,columns), dtype=np.uint8)
  
  #intervalo
  T=thres
  
  for x in range(0,rows):
    for y in range(0,columns):
      r = img[x][y]
      if T<=r:
        img2[x][y]=0
      else:
        img2[x][y]=255
  mostrar2Grises(img,img2, 'original','threshold')
  return
threshold('pastel.png',55)
```

#Seperacion por bits

```
img = cv2.imread('tree.png',cv2.IMREAD_GRAYSCALE)

lst =[]
for i in range(img.shape[0]):
  for j in range(img.shape[1]):
    lst.append(np.binary_repr(img[i][j], width=8)) # width = no. of bits

# we have a list of strings where each string represents binary pixel value
#multiply with 2^(n-1) and reshape to reconstruct the bit image

eight_bit_img= (np.array([int(i[0]) for i in lst], dtype= np.uint8)*128).reshape(img.shape[0],img.shape[1])
seven_bit_img= (np.array([int(i[1]) for i in lst], dtype= np.uint8)*64).reshape(img.shape[0],img.shape[1])
six_bit_img= (np.array([int(i[2]) for i in lst], dtype= np.uint8)*32).reshape(img.shape[0],img.shape[1])
five_bit_img= (np.array([int(i[3]) for i in lst], dtype= np.uint8)*16).reshape(img.shape[0],img.shape[1])
four_bit_img= (np.array([int(i[4]) for i in lst], dtype= np.uint8)*8).reshape(img.shape[0],img.shape[1])
three_bit_img= (np.array([int(i[5]) for i in lst], dtype= np.uint8)*4).reshape(img.shape[0],img.shape[1])
two_bit_img= (np.array([int(i[6]) for i in lst], dtype= np.uint8)*2).reshape(img.shape[0],img.shape[1])
one_bit_img= (np.array([int(i[7]) for i in lst], dtype= np.uint8)*1).reshape(img.shape[0],img.shape[1])

#concatenate these images for ease of display using cv2.hconcat()
finalr=cv2.hconcat([eight_bit_img,seven_bit_img,six_bit_img,five_bit_img])
finalv=cv2.hconcat([four_bit_img,three_bit_img,two_bit_img,one_bit_img])

#vertically concatenate
final=cv2.vconcat([finalr,finalv])
plt.imshow(final,cmap='gray')
```
